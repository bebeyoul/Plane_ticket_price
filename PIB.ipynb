# PIB annuel des pays entre 2020 et 2022

https://fr.countryeconomy.com/gouvernement/pib

```python
import urllib
import bs4
from bs4 import BeautifulSoup
from urllib import request
import requests
import pandas as pd
```

## PIB

```python
url = "https://fr.countryeconomy.com/gouvernement/pib"

#option 1
request_text = request.urlopen(url).read()
html_page = bs4.BeautifulSoup(request_text, "html")
```

```python
tableau_donnnees_PIB = html_page.find_all("div",{"class": "table-responsive"})[0]
```

```python
scrappe_pays=[]
scrappe_pib=[]
compteur=0

for tr in tableau_donnnees_PIB.findAll("tr")[1:]:
    if compteur<198 :
        # in the 1st td we get the country name
        scrappe_pays.append(tr.findAll("td")[0].text)
        # in the 3th td we get the annual PIB
        scrappe_pib.append(tr.findAll("td")[2].text)
        compteur+=1
    else:
        break
```

```python
pays=[]
for i in scrappe_pays:
    i = i[:len(i)-4]
    pays.append(i)
```

```python
pib=[]
for i in scrappe_pib:
    i = i[:len(i)-3]
    pib.append(i)
```

```python
df = pd.DataFrame(data={'Pays': pays, 'PIB annuel (M€)': pib})
df
```

```python
for i in range(len(df)):
    df["PIB annuel (M€)"][i]=df["PIB annuel (M€)"][i].replace(".","")
df
```

## PIB & Pouvoir d'achat

```python
url_2 = "https://fr.wikipedia.org/wiki/Liste_des_pays_par_PIB_(PPA)_par_habitant"

#option 1
request_text2 = request.urlopen(url_2).read()
html_page_2 = bs4.BeautifulSoup(request_text2, "html")
```

```python
PIB_par_habitant = html_page_2.find_all("table",{"class": "wikitable alternance"})[2]
```

```python
scrappe_pays_ppa=[]
scrappe_pib_ppa=[]

for tr in PIB_par_habitant.findAll("tr")[1:]:
    # in the 1st td we get the country name
    scrappe_pays_ppa.append(tr.findAll("td")[1].text)
    # in the 2th td we get the annual PIB
    scrappe_pib_ppa.append(tr.findAll("td")[2].text)
```

```python
pays_ppa=[]
for i in scrappe_pays_ppa:
    i = i.strip()
    pays_ppa.append(i)
```

```python
pib_ppa=[]
for i in scrappe_pib_ppa:
    if "(" in i:
        i = i[:len(i)-8]
    if "\n" in i:
        i = i[:len(i)-1]
    if "\xa0" in i:
        i = i.replace("\xa0", "")
    pib_ppa.append(i)
pib_ppa
```

```python
df2 = pd.DataFrame(data={'Pays': pays_ppa, 'PIB par habitant ($)': pib_ppa})
df2 = df2.astype(str)
df2
```
