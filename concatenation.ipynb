## Concatenation finale

```python
old_final_dataset = pd.read_csv("data/final_dataset.csv")
old_final_dataset= old_final_dataset.drop(columns=['Unnamed: 0'])
old_final_dataset= old_final_dataset.drop(columns=['PIB par habitant (€)'])
```

```python
# 1200 lignes
df_riche_fevr = pd.read_csv("data/A utiliser/dataset_riche_fevrier.csv")
df_riche_fevr= df_riche_fevr.drop(columns=['Unnamed: 0'])
df_riche_fevr

# 600 lignes
df_27 = pd.read_csv("data/A utiliser/df_27.csv")
# 600 lignes
df_28 = pd.read_csv("data/A utiliser/df_28.csv")
# 1200 lignes
df_26 = pd.read_csv("data/A utiliser/df_26.csv")
```

```python
# concatener les dataframes
df_new_dates = pd.concat([old_final_dataset, df_26,df_27,df_28, df_riche_fevr]).reset_index()
df_new_dates= df_new_dates.drop(columns=['index'])
```

```python
# replace / with -
df_new_dates['Date'] = df_new_dates['Date'].str.replace('/', '-')
# convertir la colonne 'date' en un format de date
df_new_dates['Date'] = pd.to_datetime(df_new_dates['Date'], format='%d-%m-%Y')
df_new_dates['Date'].unique()
df_new_dates
```

```python
# sort by date
df_new_dates = df_new_dates.sort_values(by=['Date', 'VPN']).reset_index()
df_new_dates= df_new_dates.drop(columns=['index'])
df_new_dates
```

```python
df_new_dates.to_csv('data/dataset_finale_avec_nouvelles_dates.csv')
```
