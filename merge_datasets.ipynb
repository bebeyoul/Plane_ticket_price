# Fusion les données PIB et les données des vols

```python
import pandas as pd
```

```python
# Chargement des données
df1 = pd.read_csv('data/dataset_finale_avec_nouvelles_dates.csv')
df2 = pd.read_excel('data/PIB_annuel_et_PPA_par_pays_utile.xlsx')
```

```python
# Fusion des datasets selon la colonne 'ID'
df_merged = pd.merge(df1, df2, left_on='VPN', right_on="Pays")
df_merged=df_merged.drop("Pays", axis=1)
df_merged=df_merged.drop("Unnamed: 0", axis=1)
df_merged.head()
```

```python
# convert PIB par habitant ($) into PIB par habitant (€)
df_merged['PIB par habitant (€)'] = df_merged['PIB par habitant ($)'] * 0.85
df_merged=df_merged.drop("PIB par habitant ($)", axis=1)
df_merged
```

```python
df_merged.to_csv('data/dataset_finale_avec_nouvelles_dates_et_PIB.csv')
```
